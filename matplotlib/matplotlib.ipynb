```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

# Bar Chart in matplotlib

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
import numpy as np
```

```python
xpos = np.arange(len(company))
plt.xticks(xpos, company)
plt.title("Stocks of FAANG")
plt.ylabel('in billions')
plt.xlabel('company')
plt.bar(xpos, revenue, label='Revenue')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(xpos, company)
plt.title("Stocks of FAANG")
plt.ylabel('in billions')
plt.xlabel('company')
plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.title("Stocks of FAANG")
plt.ylabel('in billions')
plt.xlabel('company')
plt.bar(xpos-0.2, revenue, label='Revenue')
plt.bar(xpos+0.2, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.title("Stocks of FAANG")
plt.ylabel('in billions')
plt.xlabel('company')
plt.bar(xpos-0.2, revenue, width = 0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width = 0.4, label='Profit')
plt.legend()
```

# for horizontal bar chart

```python
plt.yticks(xpos, company)
plt.title("Stocks of FAANG")
plt.xlabel('in billions')
plt.ylabel('company')
plt.barh(xpos-0.2,  revenue, label='Revenue')
plt.barh(xpos+0.2, profit, label='Profit')
plt.legend()
```

#  creating histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar)
```

```python
# note by default there is 10 bins or buckets in histogram
plt.hist(blood_sugar, bins=3, rwidth=0.95)
#note rwidth is relative width of bar compared to the bin size
```

```python
# custom bin size
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, color='red', histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['red', 'blue'],
        label=['men', 'women'])
plt.xlabel('Sugar level')
plt.ylabel('No of people')
plt.legend()
```

```python
# to change orientation of histogram
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['red', 'blue'],
        label=['men', 'women'], orientation='horizontal')
plt.xlabel('Sugar level')
plt.ylabel('No of people')
plt.legend()
```

# Creating pie charts in matplotlib

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

## to change radius of pie chart

```python
#by default radius is 1
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius = 2)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius = 1.5, autopct='%0.1f%%', shadow='True')
plt.show()
```

# to explode pie chart

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius = 1.5, autopct='%0.1f%%', explode=[0.1, 0, 0, 0.1, 0], startangle=45)
plt.show()
```

## saving chart in matplotlib

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius = 1.5, autopct='%0.1f%%', explode=[0.1, 0, 0, 0.1, 0])
# note we can save chart in any format like pdf, png, jpeg
plt.savefig('pie-chart.pdf', bbox_inches='tight', pad_inches=.2, orientation='landscape')
plt.show()
```
